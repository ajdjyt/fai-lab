```python
#df['num-of-doors'].replace(to_replace='two',value=2, inplace=True)
#df['num-of-doors'].replace(to_replace='four',value=4, inplace = True)
df.replace(to_replace='two',value=2, inplace=True)
df.replace(to_replace='three',value=3, inplace=True)
df.replace(to_replace='four',value=4, inplace=True)
df.replace(to_replace='five',value=5, inplace=True)
df.replace(to_replace='six',value=6, inplace=True)
df.replace(to_replace='eight',value=8, inplace=True)
df.replace(to_replace='twelve',value=12, inplace=True)
```

```python
for col in df.columns:
    a=set(type(i) for i in df[col].unique())
    if len(a)>1:
        print(col,a)
```

```python
from sklearn.preprocessing import LabelEncoder
import numpy as np
from collections import defaultdict
d = defaultdict(LabelEncoder)
def encoded(df,d):
    return df.apply(lambda x: d[x.name].fit_transform(x))
def decoded(df,d):
    return df.apply(lambda x: d[x.name].transform(x))
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.decomposition import PCA

Y = encoded(df,d)["price"]
X = encoded(df,d).drop(["price"],axis=1)
pca = PCA(n_components=5, svd_solver='full')
X_pca = pca.fit_transform(X)

x_train, x_test, y_train, y_test = train_test_split(X_pca,Y,test_size = 0.25, random_state = 42)

neigh = KNeighborsClassifier(n_neighbors=3)
neigh.fit(x_train, y_train)
```

```python
y_pred = neigh.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(y_test,y_pred)
```
